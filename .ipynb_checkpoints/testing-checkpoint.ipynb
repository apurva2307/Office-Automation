```python
def export_to_csv(df):
    main_df = pd.DataFrame()
    for table in df:

        if len(table.columns) == 8:
            if table.iloc[0][0] == 'Srno. Challanno':
                table.columns = table.iloc[0]
                table[['Sr.No.', 'Challan No.']] = table["Srno. Challanno"].apply(lambda x: pd.Series(str(x).split(" ")))
                current_df = table.iloc[:, [2, 3, 4, 8, 9]]
                current_df.columns = ["Scroll Date", "Scroll No.", "Scroll Amt", "Sr.No.", "Challan No."]
            else:
                current_df = table.iloc[:, [0, 1, 3, 4, 5]]
                current_df.columns = ["Sr.No.", "Challan No.", "Scroll Date", "Scroll No.", "Scroll Amt"]
        else:
            if table.iloc[0][0] == 'Srno. Challanno':
                table.columns = table.iloc[0]
                table[['Sr.No.', 'Challan No.']] = table["Srno. Challanno"].apply(lambda x: pd.Series(str(x).split(" ")))
                current_df = table.iloc[:, [2, 3, 4, 7, 8]]
                current_df.columns = ["Scroll Date", "Scroll No.", "Scroll Amt", "Sr.No.", "Challan No."]
            else:
                current_df = table.iloc[:, :5]
                current_df.columns = ["Sr.No.", "Challan No.", "Scroll Date", "Scroll No.", "Scroll Amt"]
        current_df = current_df[1:]
        if main_df.empty:
            main_df = current_df
        else:
            main_df = pd.concat([main_df, current_df])
        i += 1
    main_df = main_df[:-1]
    main_df.to_csv("scroll.csv", index=False)
    return main_df
```
