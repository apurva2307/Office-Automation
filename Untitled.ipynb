```python
from datetime import timedelta
import time
def match_credit_scroll(credit_scroll, debit_scroll, days=60):
    start = time.time()
    match = []
    match_num = []
    challan_dates = []
    for i in range(len(credit_scroll)):
        matchStr = ""
        matchNum = 0
        challanDates = ""
        sub_scroll = debit_scroll.loc[(debit_scroll["Challan Date"] < credit_scroll["Scroll Date"].iloc[i]) & (debit_scroll["Challan Date"] > credit_scroll["Scroll Date"].iloc[i] - timedelta(days=days))]
        for j in range(len(sub_scroll)):
            if credit_scroll["Scroll Amt"].iloc[i] == sub_scroll["Challan Amt"].iloc[j]:
                if matchStr == "":
                    matchStr = str(sub_scroll["Challan No."].iloc[j])
                else:
                    matchStr = matchStr + "; " + str(sub_scroll["Challan No."].iloc[j])
                if challanDates == "":
                    challanDates = str(sub_scroll["Challan Date"].iloc[j])
                else:
                    challanDates = challanDates + "; " + str(sub_scroll["Challan Date"].iloc[j])
                matchNum = matchNum + 1
#         print(i, end=" ")
        match.append(matchStr)
        match_num.append(matchNum)
        challan_dates.append(challanDates)
    credit_scroll["Match"] = match
    credit_scroll["Match_Num"] = match_num
    credit_scroll["Challan_Dates"] = challan_dates
    end = time.time()
    print("Time taken is ", end-start)
    return credit_scroll
```
